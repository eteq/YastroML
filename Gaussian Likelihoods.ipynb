```python
import emcee
import triangle

import numpy as np
from numpy.random import *

from matplotlib import pyplot as plt
```

```python
#interact only works for ipython 2.0 
from IPython.html.widgets import interact
have_interact = True
```

```python
def logl(x, mu, sig):
    sh = list(mu.shape)
    sh.append(1)
    mu = mu.reshape(sh)
    sig = sig.reshape(sh)
    return np.sum(-0.5*((x-mu)/sig)**2-np.log(sig),axis=-1)
```

```python
mu,sig = np.mgrid[-3:3:100j,.01:2:100j]
seed = 12345678
```

```python
figsize = (10,5)

#with uniform prior
def f(datapoints,realmu,realsig):
    r = np.random.RandomState(seed)
    data = r.randn(datapoints)*realsig + realmu
    
    like = np.exp(logl(data,mu,sig))
    likej = like/sig
    
    plt.figure(figsize=figsize)
    
    plt.subplot(1,2,1)
    plt.pcolor(mu, sig, like)
    plt.scatter(np.mean(data), np.std(data),c='w',s=10,lw=1)
    plt.xlabel(r'$\mu$')
    plt.ylabel(r'$\sigma$')
    plt.xlim(np.min(mu), np.max(mu))
    plt.ylim(np.min(sig), np.max(sig))
    plt.axvline(realmu, c='w')
    plt.axhline(realsig, c='w')
    plt.title('Uniform prior')
    
    plt.subplot(1,2,2)
    plt.pcolor(mu, sig, likej)
    plt.scatter(np.mean(data), np.std(data),c='w',s=10,lw=1)
    plt.xlabel(r'$\mu$')
    plt.ylabel(r'$\sigma$')
    plt.xlim(np.min(mu), np.max(mu))
    plt.ylim(np.min(sig), np.max(sig))
    plt.axvline(realmu, c='w')
    plt.axhline(realsig, c='w')
    plt.title('Jeffry\'s prior')
    
    plt.show()


if have_interact:
    interact(f, datapoints=(1,100), realmu=(-3,3,.1),realsig=(0.1,2,.1))
else:
    f(3)
```

```python
figsize = (5,5)

#with uniform prior
def f(npoints):
    r = np.random.RandomState(seed)
    data = r.randn(npoints)
    
    plt.figure(figsize=figsize)
    plt.hist(data, bins=np.linspace(-5,5,25))
    plt.xlim(-5,5)
    
    plt.show()

if have_interact:
    interact(f, npoints=(1,100))
else:
    f(3)
```
